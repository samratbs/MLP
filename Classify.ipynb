```
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```
# PCA 
from sklearn.decomposition import PCA
nof_prin_components = 195 # PARAMETER for optimisation in expereiments
pca = PCA(n_components=nof_prin_components, whiten=True).fit(X_train)
# applies PCA to the train and test images to calculate the principal components
X_train_pca = pca.transform(X_train) 
X_test_pca = pca.transform(X_test) 
```

```
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
svc=SVC(probability=True, kernel='linear')
ada_clf = AdaBoostClassifier(base_estimator=svc, n_estimators=428,
    algorithm="SAMME.R", learning_rate=0.01, random_state=42)
# ada_clf = AdaBoostClassifier(
#     DecisionTreeClassifier(max_depth=3), n_estimators=428,
#     algorithm="SAMME.R", learning_rate=0.0101, random_state=42)
```

```
ada_clf.get_params()
```

```
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, truncnorm, randint
dist = dict(learning_rate=uniform(0.001,0.01), n_estimators=randint(0,500))
```

```
clf = RandomizedSearchCV(estimator=ada_clf, param_distributions=dist, n_iter=10)
search = clf.fit(X_train_pca, y_train)
search.best_params_
```

```
y_pred = ada_clf.predict(X_test_pca)
```

```
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, y_pred))
```
